```python
from sklearn.linear_model import LogisticRegression            #é€»è¾‘å›žå½’
from sklearn.neighbors import KNeighborsClassifier             #Kä¸´è¿‘åˆ†ç±»
from sklearn import svm                                        #æ”¯æŒå‘é‡æœº
from sklearn import metrics                                    #for checking the model accuracy    
from sklearn.tree import DecisionTreeClassifier                #å†³ç­–æ ‘
```

```python
from sklearn.model_selection import train_test_split 
```

```python
train, test = train_test_split(data, test_size = 0.3)
```

```python
train.shape
```

```python
test.shape
```

```python
train_X = train[['a', 'b', 'c', 'd','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v']]
train_y = train['w']
test_X = test[['a', 'b', 'c', 'd','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v']]
test_y = test['w']

train_X.head(2)
```

```python
model = svm.SVC()                            
model.fit(train_X,train_y)                   
prediction = model.predict(test_X)           
metrics.accuracy_score(prediction,test_y)
```
